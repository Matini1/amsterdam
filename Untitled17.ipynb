```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Function to plot scatter plots
def plotter(data, a, b):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=a, y='Price', data=data)
    plt.title(f'Scatter plot between Price and {a}')
    plt.xlabel(a)
    plt.ylabel('Price')
    plt.show()

# Load the data
data = pd.read_csv('HousingPrices-Amsterdam-August-2021.csv')

# Suppress warnings
warnings.filterwarnings('ignore')

# Plotting different combinations
plotter(data, 'Price', 'Address')
plotter(data, 'Area', 'Room')
plotter(data, 'Lat', 'Lon')
plotter(data, 'Zip', 'Unnamed: 0')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
data = pd.read_csv('HousingPrices-Amsterdam-August-2021.csv')

# Call the plotter function with the desired columns
plotter(data, 'Room', 'Zip')
plotter(data, 'Address', 'Price')
plotter(data, 'Lon', 'Area')
plotter(data, 'lat', 'price')

# Show the plots
plt.show()
```

```python
x = data.drop('Price', axis = 1)
y = data.Price


from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=2, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# Drop non-numeric columns for now
xtrain_numeric = xtrain.drop(['Address'], axis=1)
xtest_numeric = xtest.drop(['Address'], axis=1)

# Combine train and test data for encoding
combined_data = pd.concat([xtrain_numeric, xtest_numeric], ignore_index=True)

# Use LabelEncoder for 'Zip' column
label_encoder = LabelEncoder()
combined_data['Zip'] = label_encoder.fit_transform(combined_data['Zip'])

# One-hot encode other categorical columns if needed
# For example, if 'Room' is categorical:
# combined_data = pd.get_dummies(combined_data, columns=['Room'])

# Split back into train and test
xtrain_encoded = combined_data[:len(xtrain_numeric)]
xtest_encoded = combined_data[len(xtrain_numeric):]

# Handle missing values
imputer = SimpleImputer(strategy='median')
xtrain_imputed = imputer.fit_transform(xtrain_encoded)
xtest_imputed = imputer.transform(xtest_encoded)

# Check if there are any missing values in ytrain
missing_ytrain = ytrain.isnull().sum()
if missing_ytrain > 0:
    # Fill missing values with median
    ytrain = ytrain.fillna(ytrain.median())

# Initialize the RandomForestRegressor
rf_regressor = RandomForestRegressor()

# Train the model
rf_regressor.fit(xtrain_imputed, ytrain)
validation = rf_regressor.predict(xtrain_imputed)
print("R2 Score on Training Data:", r2_score(ytrain, validation) * 100)

# Evaluate on test data
test_predictions = rf_regressor.predict(xtest_imputed)
print("R2 Score on Test Data:", r2_score(ytest, test_predictions) * 100)
```
